# Georeferencing Audit Analysis

This notebook performs an audit of georeferencing data using two distinct methods to identify inconsistencies in location data, such as those for client addresses. The analysis is designed to be generic, reusable, and suitable for public sharing on GitHub, using synthetic data to ensure privacy.

## Objectives
- **Method 1: Real Coordinates Analysis**: Identify clusters of reference (ground truth) coordinates that are too close to each other, indicating potential duplicates or data entry errors.
- **Method 2: Estimated Coordinates Analysis**: Compare reference coordinates with coordinates estimated via a geocoding API (e.g., Google Maps API) to detect inaccuracies in georeferencing.
- Provide statistical insights and visualizations for both methods to assess data quality.
- Offer actionable recommendations for improving georeferencing processes.

## Structure
1. **Setup and Dependencies**: Install and import required libraries.
2. **Data Preparation**: Generate synthetic data and simulate API-based geocoding.
3. **Method 1: Real Coordinates Analysis**: Analyze distances between reference coordinates.
4. **Method 2: Estimated Coordinates Analysis**: Compare reference vs. estimated coordinates.
5. **Error Distribution Analysis**: Visualize error distributions for both methods.
6. **Advanced Visualizations**: Create interactive maps and plots for deeper insights.
7. **Results Export**: Save results to CSV files.

## 1. Setup and Dependencies

Install and import libraries for data processing, geospatial calculations, and visualization.

```python
# Install required libraries (uncomment if needed)
# !pip install pandas numpy matplotlib seaborn geopy plotly

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from geopy.distance import geodesic
import plotly.express as px
import plotly.graph_objects as go
from datetime import datetime

# Set visualization style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)
```

## 2. Data Preparation

We generate a synthetic dataset to simulate georeferencing data for client locations. The dataset includes:
- `ID`: Unique identifier.
- `Name`: Generic entity name (e.g., 'Location_001').
- `Address`: Generic address (e.g., 'Street_001').
- `Lat_ref`, `Lon_ref`: Reference (ground truth) coordinates.
- `Lat_est`, `Lon_est`: Estimated coordinates (simulating API geocoding with noise).

For **Method 2**, the estimated coordinates mimic the output of a geocoding API (e.g., Google Maps) by introducing realistic errors. In a real implementation, you would replace the simulated geocoding with an API call (see README for details).

```python
def generate_synthetic_data(n_records=100, lat_center=40.7128, lon_center=-74.0060, spread=0.05, max_error=0.01):
    """Generate synthetic georeferencing data with simulated geocoding errors.

    Args:
        n_records (int): Number of records.
        lat_center (float): Central latitude (e.g., New York City).
        lon_center (float): Central longitude.
        spread (float): Spread of points around the center (degrees).
        max_error (float): Maximum error for estimated coordinates (degrees).

    Returns:
        pd.DataFrame: Synthetic dataset.
    """
    np.random.seed(42)
    data = {
        'ID': [f'LOC_{i:03d}' for i in range(n_records)],
        'Name': [f'Location_{i:03d}' for i in range(n_records)],
        'Address': [f'Street_{i:03d}, City' for i in range(n_records)],
        'Lat_ref': lat_center + np.random.uniform(-spread, spread, n_records),
        'Lon_ref': lon_center + np.random.uniform(-spread, spread, n_records),
    }
    df = pd.DataFrame(data)
    # Simulate geocoding errors
    df['Lat_est'] = df['Lat_ref'] + np.random.uniform(-max_error, max_error, n_records)
    df['Lon_est'] = df['Lon_ref'] + np.random.uniform(-max_error, max_error, n_records)
    return df

# Generate synthetic data
df = generate_synthetic_data(n_records=100)

# Validate data
if df.isnull().any().any():
    raise ValueError('Generated data contains missing values.')

print('Synthetic Data Sample:')
df.head()
```

## 3. Method 1: Real Coordinates Analysis

This method compares reference coordinates among themselves to identify pairs that are closer than a threshold (e.g., 500 meters). Close proximity may indicate duplicates or errors, as distinct locations (e.g., client addresses) should not be too close.

```python
def find_close_pairs(df, threshold_km=0.5, lat_col='Lat_ref', lon_col='Lon_ref'):
    """Identify pairs of records with reference coordinates closer than the threshold.

    Args:
        df (pd.DataFrame): DataFrame with coordinate data.
        threshold_km (float): Distance threshold in kilometers.
        lat_col (str): Column name for latitude.
        lon_col (str): Column name for longitude.

    Returns:
        pd.DataFrame: DataFrame with close pairs and distances.
    """
    close_pairs = []
    for i, row1 in df.iterrows():
        for j, row2 in df.iterrows():
            if i < j:
                coord1 = (row1[lat_col], row1[lon_col])
                coord2 = (row2[lat_col], row2[lon_col])
                try:
                    distance_km = geodesic(coord1, coord2).kilometers
                    if distance_km < threshold_km:
                        close_pairs.append({
                            'ID_1': row1['ID'],
                            'Name_1': row1['Name'],
                            'Address_1': row1['Address'],
                            'ID_2': row2['ID'],
                            'Name_2': row2['Name'],
                            'Address_2': row2['Address'],
                            'Distance_km': distance_km
                        })
                except (ValueError, TypeError):
                    continue
    return pd.DataFrame(close_pairs)

# Analyze close pairs
close_pairs_df = find_close_pairs(df, threshold_km=0.5)

# Display and save results
if not close_pairs_df.empty:
    print('Close Pairs Found (within 500 meters):')
    print(close_pairs_df)
    close_pairs_df.to_csv('close_pairs_real.csv', index=False)
    print('Close pairs saved to close_pairs_real.csv')
else:
    print('No pairs found within 500 meters.')
```

## 4. Method 2: Estimated Coordinates Analysis

This method compares reference coordinates with estimated coordinates (simulating geocoding API output) to quantify georeferencing errors. In a real implementation, you would use an API like Google Maps to geocode addresses.

```python
def calculate_distance(row, lat_ref_col='Lat_ref', lon_ref_col='Lon_ref', lat_est_col='Lat_est', lon_est_col='Lon_est'):
    """Calculate geodesic distance between reference and estimated coordinates.

    Args:
        row (pd.Series): DataFrame row with coordinate columns.
        lat_ref_col (str): Column name for reference latitude.
        lon_ref_col (str): Column name for reference longitude.
        lat_est_col (str): Column name for estimated latitude.
        lon_est_col (str): Column name for estimated longitude.

    Returns:
        float: Distance in meters, or NaN if coordinates are invalid.
    """
    try:
        point_ref = (row[lat_ref_col], row[lon_ref_col])
        point_est = (row[lat_est_col], row[lon_est_col])
        return geodesic(point_ref, point_est).meters
    except (ValueError, TypeError):
        return np.nan

# Calculate distances
df['Distance_m'] = df.apply(calculate_distance, axis=1)

# Handle invalid distances
if df['Distance_m'].isnull().any():
    print('Warning: Some distances could not be calculated. Dropping invalid records.')
    df = df.dropna(subset=['Distance_m'])

# Basic statistics
distance_stats = df['Distance_m'].describe()
print('Distance Statistics (meters):')
print(distance_stats)

# Save distances
df[['ID', 'Name', 'Address', 'Distance_m']].to_csv('georeferencing_distances_est.csv', index=False)
print('Distances saved to georeferencing_distances_est.csv')
```

## 5. Error Distribution Analysis

Analyze the distribution of errors for both methods:
- **Method 1**: Distribution of distances between close pairs (if any).
- **Method 2**: Distribution of errors between reference and estimated coordinates.

```python
# Method 1: Distribution of close pairs (if any)
if not close_pairs_df.empty:
    plt.figure(figsize=(10, 6))
    sns.histplot(close_pairs_df['Distance_km'], bins=20, color='skyblue')
    plt.xlabel('Distance (kilometers)')
    plt.ylabel('Number of Pairs')
    plt.title('Distribution of Distances Between Close Pairs (Method 1)')
    plt.tight_layout()
    plt.savefig('close_pairs_distribution.png')
    plt.show()

    # Interactive Plotly histogram
    fig = px.histogram(close_pairs_df, x='Distance_km', nbins=20,
                       title='Close Pairs Distance Distribution (Method 1)',
                       labels={'Distance_km': 'Distance (km)', 'count': 'Number of Pairs'},
                       template='plotly_white')
    fig.update_layout(bargap=0.1)
    fig.write_html('close_pairs_histogram.html')
    print('Close pairs distribution saved as close_pairs_histogram.html')

# Method 2: Distribution of georeferencing errors
bins = [0, 100, 500, 2000, float('inf')]
labels = ['0-100m', '100-500m', '500-2000m', '2000m+']
df['Error_Range'] = pd.cut(df['Distance_m'], bins=bins, labels=labels, include_lowest=True)
error_dist = df['Error_Range'].value_counts().sort_index()

# Matplotlib bar plot
plt.figure(figsize=(10, 6))
error_dist.plot(kind='bar', color='lightcoral', edgecolor='black')
plt.xlabel('Error Range (meters)')
plt.ylabel('Number of Records')
plt.title('Distribution of Georeferencing Errors (Method 2)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('error_distribution_est.png')
plt.show()

# Interactive Plotly histogram
fig = px.histogram(df, x='Distance_m', nbins=30,
                   title='Georeferencing Error Distribution (Method 2)',
                   labels={'Distance_m': 'Error (meters)', 'count': 'Number of Records'},
                   template='plotly_white')
fig.update_layout(bargap=0.1)
fig.write_html('error_histogram_est.html')
print('Error distribution saved as error_histogram_est.html')

# Save distributions
if not close_pairs_df.empty:
    close_pairs_df['Distance_km'].describe().to_csv('close_pairs_stats.csv')
    print('Close pairs statistics saved to close_pairs_stats.csv')
error_dist.to_csv('error_distribution_est.csv')
print('Error distribution saved to error_distribution_est.csv')
```

## 6. Advanced Visualizations

Create interactive maps and scatter plots to visualize:
- **Method 1**: Locations of close pairs to highlight potential duplicates.
- **Method 2**: Reference vs. estimated coordinates to show error magnitudes.

```python
# Method 1: Map of close pairs
if not close_pairs_df.empty:
    close_points = df[df['ID'].isin(close_pairs_df['ID_1']) | df['ID'].isin(close_pairs_df['ID_2'])]
    fig = px.scatter_map(close_points, lat='Lat_ref', lon='Lon_ref', hover_data=['ID', 'Name', 'Address'],
                            color_discrete_sequence=['red'], size_max=10,
                            title='Close Pairs Locations (Method 1)')
    fig.update_layout(mapbox_style='open-street-map',
                      mapbox_center={'lat': df['Lat_ref'].mean(), 'lon': df['Lon_ref'].mean()},
                      mapbox_zoom=10, height=600)
    fig.write_html('close_pairs_map.html')
    print('Close pairs map saved as close_pairs_map.html')

# Method 2: Reference vs. Estimated Coordinates
fig = go.Figure()
fig.add_trace(go.Scattermap(
    lat=df['Lat_ref'], lon=df['Lon_ref'],
    mode='markers', marker=dict(size=8, color='blue', opacity=0.6),
    text=df['ID'], name='Reference Coordinates'
))
fig.add_trace(go.Scattermap(
    lat=df['Lat_est'], lon=df['Lon_est'],
    mode='markers', marker=dict(size=8, color='red', opacity=0.6),
    text=df['ID'], name='Estimated Coordinates'
))
fig.update_layout(
    title='Reference vs. Estimated Coordinates (Method 2)',
    mapbox=dict(style='open-street-map', center=dict(lat=df['Lat_ref'].mean(), lon=df['Lon_ref'].mean()), zoom=10),
    showlegend=True, height=600
)
fig.write_html('coordinates_map_est.html')
print('Coordinates map saved as coordinates_map_est.html')

# Error scatter plot (Method 2)
fig = px.scatter(df, x='Lon_ref', y='Lat_ref', size='Distance_m', color='Distance_m',
                 hover_data=['ID', 'Name', 'Address'],
                 title='Georeferencing Errors by Location (Method 2)',
                 labels={'Lon_ref': 'Longitude', 'Lat_ref': 'Latitude', 'Distance_m': 'Error (meters)'},
                 color_continuous_scale='Viridis')
fig.update_layout(template='plotly_white')
fig.write_html('error_scatter_est.html')
print('Error scatter plot saved as error_scatter_est.html')
```

## 7. Results Export

Export results with timestamps to maintain version control.

```python
timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')

# Full dataset
df.to_csv(f'georeferencing_audit_full_{timestamp}.csv', index=False)
print(f'Full dataset saved to georeferencing_audit_full_{timestamp}.csv')

# Top/bottom errors (Method 2)
top_errors = df.nlargest(10, 'Distance_m')[['ID', 'Name', 'Address', 'Distance_m']]
bottom_errors = df.nsmallest(10, 'Distance_m')[['ID', 'Name', 'Address', 'Distance_m']]
top_errors.to_csv(f'top_errors_est_{timestamp}.csv', index=False)
bottom_errors.to_csv(f'bottom_errors_est_{timestamp}.csv', index=False)
print(f'Top/bottom errors saved to top_errors_est_{timestamp}.csv and bottom_errors_est_{timestamp}.csv')
```
